```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import constants as cts
import math as m
%matplotlib inline
```

```python
x=np.linspace(-1,1,100)
g=cts.gravitational_constant
m1=1.6e26
m2=1.6e26
h=10000
mu=g*(m1+m2)
alpha=(3/2)*(h**2/mu)
E=0
beta=(3/mu)*E
f=x**3-alpha*x**2+beta
plt.plot(x,f)
plt.grid()
plt.show()
```

```python
x=np.linspace(0,2*np.pi,100)
g=cts.gravitational_constant
m1=1.6e26
m2=1.6e26
h=10000
mu=g*(m1+m2)
r=(mu/(3*h**2))*(np.cos(x)+1)
plt.polar(x,r)
plt.title("Movimiento para H=0 \n")
plt.savefig("Cardiod.png")
plt.show()
```

Energía negativa

```python
x=np.linspace(-1,1,100)
g=cts.gravitational_constant
m1=1.6e26
m2=1.6e26
h=10000
mu=g*(m1+m2)
alpha=(3./2.)*(h**2/mu)
beta=(4./27.)*alpha**3
f=x**3-alpha*x**2+beta
plt.plot(x,f)
plt.ylim(-.2,0.2)
plt.xlim(-.5,0.5)
plt.grid()
plt.show()
```

```python
x=np.linspace(0,20,1000)
g=cts.gravitational_constant
m1=1.6e26
m2=1.6e26
h=10000
mu=g*(m1+m2)
r=(mu/(h**2))+x*0
r2=[]
for i in x:
    r2.append((mu/(3*h**2))*(m.cosh(i)-1))
plt.plot(x,r,x,r2)
plt.title("radio \n")
plt.show()
```

```python
x=np.linspace(0,10,100)
g=cts.gravitational_constant
m1=1.6e26
m2=1.6e26
h=10000
mu=g*(m1+m2)
r=(mu/(h**2))+x*0
r2=[]
for i in x:
    r2.append((mu/(h**2))*((m.cosh(i)-1)/(m.cosh(i)+2)))
plt.plot(x,r,x,r2)
plt.title("radio \n")
plt.show()
```
